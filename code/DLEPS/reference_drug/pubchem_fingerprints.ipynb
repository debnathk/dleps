# Calculating Pubchem Fingerprints:
- For Reference drugs
- For Training drugs

```python
import pandas as pd
```

## Pubchem fingerprints calculation: Reference Drugs

```python
df = pd.read_csv('gdsc_drugs+SMILES.csv')
df
```

```python
smiles_list = list(df['standardized_SMILES'])
smiles_list
```

```python
# only calculate fingerprints
from padelpy import from_smiles
fingerprints = from_smiles(smiles_list, fingerprints=True, descriptors=False, threads=10, output_csv='pubchem_fp_ref.csv')
```

```python
df_fp = pd.read_csv('pubchem_fp_ref.csv')
df_fp.head()
```

```python
df_fp = df_fp.drop(columns='Name')
df_fp.insert(0, 'DRUG_NAME', list(df['DRUG_NAME']))
df_fp.to_csv('pubchem_fp_ref.csv', index=False)
df_fp.head()
```
